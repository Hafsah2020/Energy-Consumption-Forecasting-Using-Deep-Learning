```python
import pandas as pd
import numpy as np
```

```python
def save(data, sheetname, append = True):
    if append:
        with pd.ExcelWriter('../data/processed_data/Energy_consumption.xlsx', mode = 'a', engine = 'openpyxl' ) as writer:
            data.to_excel(writer, sheet_name = sheetname)
    else:
        with pd.ExcelWriter('../data/processed_data/Energy_consumption.xlsx') as writer:
            data.to_excel(writer, sheet_name = sheetname)
```

```python
data_19 = pd.read_excel('../data/raw_data/2019 Power sector Energy Generated.xlsx', sheet_name = 'Energy Consumed (MWH) 2019', header = 1, index_col = 0)
```

```python
data_19 = data_19.drop([np.nan, 'JAN-JUNE TOTAL', 'JUL-DEC TOTAL'])
```

```python
data_19
```

```python
save(data_19, 'Consumption in 2019', False)
```
